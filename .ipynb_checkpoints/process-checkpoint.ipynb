## Crude oil

```python
df = pd.read_excel("data.xlsx",1)
```

```python
df = df.iloc[2:]
```

```python
df.columns = df.iloc[0].values
```

```python
df = df.iloc[1:]
```

```python
columns = df.columns.values
columns[0] = "Date"
df.columns = columns
```

```python
df["month"] = df.Date.dt.month
df["year"] = df.Date.dt.year
```

```python
df.drop(columns="Date",inplace=True)
```

```python
df.Open = df.Open.astype(float)
```

```python
df.High = df.High.astype(float)
df.Low = df.Low.astype(float)
df.Close = df.Close.astype(float)
```

```python
df = df.groupby(["month","year"]).agg({"Close":np.mean,"High":np.max,"Low":np.min}).reset_index()
```

```python
df.to_csv("clean_crude.csv",index=None)
```

```python
def split_date_time(dt):

    try:

        date, time = dt.split()

    except (ValueError, AttributeError):

        return float('nan')

    try:

        month, day, year = date.split('/')

    except (ValueError, AttributeError):

        return float('nan')

    try:

        hour, minute = time.split(':')

    except (ValueError, AttributeError):

        return float('nan')

    return pd.Timestamp(year=int(year), month=int(month), day=int(day), hour=int(hour), minute=int(minute))
```
